Removing Columns

```python
df1 = df.copy()
```

```python
df2=df1.drop(columns=['Postal Code', 'Ship Mode']) 
```

Which state had the highest / lowest profit. 

1. create a variable 'Grouped' where the DF is grouped by States that include sum of sales.
2. create another variable 'Sorted_grouped' that sorts the states by highest to lowest Sales.

```python
grouped_states=df2.groupby(by=[ 'State']).sum()
```

```python
sorted_grouped = grouped_states.sort_values(by='Sales',ascending=False)
```

```python
sorted_grouped
#Highest: California  & Lowest : California & North Dakota
```

Which category has the highest Profit? 
1. create a variable 'Grouped' where the DF is grouped by category that include sum of profit/sales.
2. create another variable 'Sorted_category' that sorts the category by highest to lowest profit/sales.

```python
grouped_categories = df2.groupby(by=[ 'Category']).sum()
```

```python
grouped_categories
```

```python
sorted_categories = grouped_categories.sort_values(by=  'Profit', ascending=False)
```
